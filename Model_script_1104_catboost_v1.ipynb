```python
import numpy as np
import pandas as pd
import joblib

from catboost import CatBoostClassifier

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import FunctionTransformer, StandardScaler,OneHotEncoder

#from typing import List, Optional
#from feature_selection import DataSelector

import pickle
```

```python
train = pd.read_csv('/Users/olga/Documents/datasets/training data/Lending club/in/10K_Lending_Club_Loans_train.csv')
y = train['is_bad']
X= train.drop(['is_bad','earliest_cr_line','url'], axis=1)
```

```python
#X.dtypes
```

```python
#X[cat_features]
```

```python
# Fill null values for Categorical Features
cat_features = ['term', 'int_rate', 'grade', 'sub_grade', 'emp_length', 'emp_title','desc','purpose','title','home_ownership', 'verification_status', 'pymnt_plan', 'loan_id', 'zip_code', 'addr_state', 'initial_list_status']
text_features = ['emp_title', 'url', 'desc', 'purpose', 'title']

for c in cat_features:
    X[c] = X[c].fillna('unknown')
    try:
        X[c] = X[c].astype(str)
    except:
        pass

    
# Fill null values for numerical Features
X = X.fillna(0)
```

```python
model = CatBoostClassifier(iterations=10,
                           depth=2,
                           learning_rate=1,
                           loss_function='Logloss',
                            verbose=True)
```

```python
# train the model
model.fit(X,y, cat_features)
```

```python
test_data = pd.read_csv('/Users/olga/Documents/datasets/training data/Lending club/in/10K_Lending_Club_test.csv')
X_test= test_data.drop(['earliest_cr_line','url'], axis=1)
for c in cat_features:
    X_test[c] = X_test[c].fillna('unknown')
    try:
        X_test[c] = X_test[c].astype(str)
    except:
        pass

X_test = X_test.fillna(0)
```

```python
model.predict_proba(X_test)
```
